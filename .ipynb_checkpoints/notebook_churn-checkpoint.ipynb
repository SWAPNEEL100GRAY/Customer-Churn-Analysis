## 4 - Customer Churn Prediction

After completing the exploratory data analysis, the next step is to build machine learning models that can predict whether a customer is likely to churn. This section uses the cleaned customer data to train and evaluate classification models.


### 4.1 Preparing the Data for Machine Learning

For machine learning, the dataset is loaded again so the prediction workflow is independent from the EDA cells above. The `TotalCharges` column is converted from text to numeric, missing values are removed, and `customerID` is dropped because it is only an identifier and does not help predict churn.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

ml_df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

ml_df["TotalCharges"] = pd.to_numeric(ml_df["TotalCharges"], errors="coerce")
ml_df = ml_df.dropna(subset=["TotalCharges"]).copy()
ml_df = ml_df.drop(columns=["customerID"])

X = ml_df.drop("Churn", axis=1)
y = ml_df["Churn"].map({"No": 0, "Yes": 1})

print("Feature matrix shape:", X.shape)
print("Target distribution:")
print(y.value_counts(normalize=True).rename("proportion"))
```

### 4.2 Train/Test Split

The data is split into training and testing sets. The model learns patterns from the training set, while the test set is kept separate to evaluate how well the model performs on unseen customer records. Stratification is used so both sets keep a similar churn/non-churn distribution.


```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

### 4.3 One-Hot Encoding

Most customer attributes are categorical, such as contract type, payment method, and internet service. Machine learning models require numerical input, so categorical columns are converted into one-hot encoded columns. Numerical columns are kept numeric, and scaling is applied for Logistic Regression.


```python
categorical_features = X_train.select_dtypes(include=["object", "string"]).columns.tolist()
numerical_features = X_train.select_dtypes(exclude=["object", "string"]).columns.tolist()

logistic_preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)

forest_preprocessor = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numerical_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)

print("Categorical features:", categorical_features)
print("Numerical features:", numerical_features)
```

### 4.4 Model Evaluation Function

The same evaluation process is used for both models. Accuracy shows overall correctness, precision shows how many predicted churners actually churned, recall shows how many actual churners were detected, F1 score balances precision and recall, and ROC-AUC measures how well the model separates churners from non-churners.


```python
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    metrics = {
        "Model": model_name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_proba)
    }

    print(model_name)
    for metric, value in metrics.items():
        if metric != "Model":
            print(f"{metric}: {value:.4f}")

    cm = confusion_matrix(y_test, y_pred)
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Churn", "Churn"])
    display.plot(cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

    return metrics
```

### 4.5 Logistic Regression

Logistic Regression is used as the baseline model because it is simple, interpretable, and commonly used for binary classification problems such as churn prediction.


```python
logistic_model = Pipeline(
    steps=[
        ("preprocessor", logistic_preprocessor),
        ("model", LogisticRegression(max_iter=1000, random_state=42))
    ]
)

logistic_model.fit(X_train, y_train)
logistic_results = evaluate_model(logistic_model, X_test, y_test, "Logistic Regression")
```

#### Logistic Regression Interpretation

Logistic Regression provides a strong baseline for this churn prediction task. It usually performs well when churn has a mostly linear relationship with customer attributes such as tenure, contract type, monthly charges, and payment method. In this run, the model balances overall accuracy with a reasonable ability to detect churners, making it useful for an initial retention targeting strategy.


### 4.6 Random Forest Classifier

Random Forest is used as a second model because it can capture non-linear relationships and interactions between features. It combines many decision trees to produce a more robust prediction than a single tree.


```python
random_forest_model = Pipeline(
    steps=[
        ("preprocessor", forest_preprocessor),
        ("model", RandomForestClassifier(
            n_estimators=300,
            random_state=42,
            class_weight="balanced"
        ))
    ]
)

random_forest_model.fit(X_train, y_train)
random_forest_results = evaluate_model(random_forest_model, X_test, y_test, "Random Forest Classifier")
```

#### Random Forest Interpretation

The Random Forest model can identify more complex patterns in customer behavior, but it may not always outperform Logistic Regression on this dataset. If Random Forest has lower recall, it means the model misses more actual churn customers. For churn prevention, recall is especially important because the business wants to identify as many at-risk customers as possible before they leave.


### 4.7 Model Comparison

The final step is to compare the models side by side. For churn prediction, accuracy alone is not enough because the dataset has more non-churn customers than churn customers. Recall, F1 Score, and ROC-AUC are especially important when deciding which model is more useful for customer retention.


```python
model_comparison = pd.DataFrame([logistic_results, random_forest_results])
model_comparison = model_comparison.set_index("Model")
model_comparison.round(4)
```

### Final Machine Learning Conclusion

Both models can be used to predict customer churn, but the better model should be selected based on business priorities. If the company wants a simple and interpretable model, Logistic Regression is a strong choice. If the company wants to capture more complex customer behavior, Random Forest can be explored further with hyperparameter tuning.

For a churn reduction use case, recall should receive special attention because failing to identify a customer who is likely to churn may result in lost revenue. Future improvements could include hyperparameter tuning, cross-validation, feature importance analysis, and threshold adjustment to improve churn detection.


## 5 - Feature Importance Analysis

Feature importance explains which input variables had the strongest influence on the Random Forest model's churn predictions. In a Random Forest, features that create better splits across many decision trees receive higher importance scores. This helps connect the machine learning model back to business understanding.


### 5.1 Extracting Feature Importances

The Random Forest model was trained inside a pipeline with one-hot encoding. Because categorical variables were expanded into multiple encoded columns, the feature names are extracted from the preprocessing step before ranking their importance scores.


```python
encoded_feature_names = random_forest_model.named_steps["preprocessor"].get_feature_names_out()
forest_importances = random_forest_model.named_steps["model"].feature_importances_

feature_importance_df = pd.DataFrame({
    "Feature": encoded_feature_names,
    "Importance": forest_importances
}).sort_values(by="Importance", ascending=False)

feature_importance_df["Feature"] = (
    feature_importance_df["Feature"]
    .str.replace("num__", "", regex=False)
    .str.replace("cat__", "", regex=False)
)

top_10_features = feature_importance_df.head(10).copy()
top_10_features
```

### 5.2 Top 10 Important Features

The table above ranks the features that contributed most to the Random Forest model's churn predictions. A higher importance value means the feature was more useful for separating customers who churned from customers who stayed.


```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_10_features,
    x="Importance",
    y="Feature",
    palette="viridis"
)
plt.title("Top 10 Feature Importances - Random Forest")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

### 5.3 Why These Features Matter

The most important features usually represent customer commitment, pricing pressure, service experience, and payment behavior. These areas are closely connected to churn because they influence how much value customers feel they are receiving from the telecom service.


```python
def explain_feature(feature):
    if feature == "tenure":
        return "Tenure reflects customer loyalty. Newer customers often have weaker attachment to the service and are more likely to churn."
    if feature == "MonthlyCharges":
        return "Monthly charges reflect pricing pressure. Higher bills can increase churn when customers do not feel they receive enough value."
    if feature == "TotalCharges":
        return "Total charges combine tenure and spending. They help separate long-term retained customers from newer or lower-value customers."
    if "Contract_Month-to-month" in feature:
        return "Month-to-month customers have low switching barriers, so they can leave more easily than customers on annual contracts."
    if "Contract_Two year" in feature or "Contract_One year" in feature:
        return "Long-term contracts usually indicate stronger commitment and reduce the likelihood of churn."
    if "PaymentMethod_Electronic check" in feature:
        return "Electronic check customers often show higher churn risk, possibly due to weaker payment commitment or different customer preferences."
    if "InternetService_Fiber optic" in feature:
        return "Fiber optic customers may churn when higher prices or service expectations are not matched by perceived value."
    if "OnlineSecurity_No" in feature:
        return "Customers without online security may be less engaged with bundled services and therefore easier to lose."
    if "TechSupport_No" in feature:
        return "Customers without tech support may have unresolved service issues, increasing frustration and churn risk."
    if "PaperlessBilling_Yes" in feature:
        return "Paperless billing can be associated with more price-aware or digitally active customers who compare alternatives easily."
    if "SeniorCitizen" in feature:
        return "Senior citizen status captures a demographic segment that may respond differently to pricing, support, and service complexity."
    if "Dependents_No" in feature or "Partner_No" in feature:
        return "Customers without household ties may have fewer shared service needs, making switching easier."
    return "This feature helps the model identify behavioral or service patterns associated with customer churn."

top_10_explanations = top_10_features.copy()
top_10_explanations["Business Meaning"] = top_10_explanations["Feature"].apply(explain_feature)
top_10_explanations
```

### 5.4 Business Interpretation

Feature importance turns the model from a prediction tool into a decision-support tool. If the most important features are related to contract type, tenure, monthly charges, payment method, or support services, the business should focus retention efforts on those areas.

For example, customers with short tenure and month-to-month contracts may need stronger onboarding, early engagement, and upgrade incentives. Customers with high monthly charges may need better value communication, discounts, or bundled offers. Customers without support or security services may benefit from targeted service add-ons that increase engagement and reduce churn risk.


## Top Drivers of Customer Churn

The key factors influencing customer churn and retention are customer tenure, contract commitment, monthly cost, total spending, payment behavior, internet service type, and support-related services. Customers are more likely to churn when they are newer, have flexible month-to-month contracts, pay higher monthly charges, use riskier payment methods, or lack support and security services.

To improve retention, the company should prioritize early-lifecycle customers, encourage longer-term contracts, improve perceived value for high-bill customers, and offer support-focused bundles to customers with fewer protective services.
